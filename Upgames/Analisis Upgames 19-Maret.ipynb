```python
import numpy as pd
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
df = pd.read_csv('transaksi_valid')

df['lt'].fillna(value=1, inplace=True)
# df.pop('id')
df.head()
```

```python
sns.set_style('darkgrid')
```

```python
# Ganti format waktu
df['Waktu'] = df['tgl'] + ' ' + df['jam']
df['Waktu'] = pd.to_datetime(df['Waktu'])
```

```python
df.pop('jam')
```

```python
# df.pop('tgl')
```

```python
df.head()
```

```python
mappaket = {
    1:'Billyard',
    2:'PS4',
    3:'PC',
    4:'PS3'
}
df['jenis'] = df['jenis'].map(mappaket)
```

```python
df.head()
```

```python
df['Jam'] = df['Waktu'].apply(lambda time: time.hour)
df['Bulan'] = df['Waktu'].apply(lambda time: time.month)
df['Hari'] = df['Waktu'].apply(lambda time: time.dayofweek)
# df['tgl'] = df['Waktu'].apply(lambda time: time.date)
```

```python
# Ganti format hari jadi nama hari
maphari = {0:'Senin',1:'Selasa',2:'Rabu',3:'Kamis',4:'Jumat',5:'Sabtu',6:'Minggu'}

df['Hari'] = df['Hari'].map(maphari)
```

```python
df.head()
```

```python
# df['nama_paket'].value_counts()
```

```python
def convert_to_string(x):
    return str(x)
```

```python
# df.head()


df['nama_paket'] = df['nama_paket'].apply(lambda x : convert_to_string(x))
df['paket_valid'] = df['nama_paket'].apply(lambda x : x.split()[0])
df.drop(df[df['paket_valid'] != 'PAKET'].index, inplace=True) #MENGHAPUS VALUE ERROR nama_paket
```

```python
# df.pop('paket_valid')
```

```python
# JUMLAH JENIS TERBANYAK TIAP HARINYA

plt.figure(figsize=(12,7))
sns.countplot(x='Hari',data=df,hue='jenis',palette='colorblind')

# To relocate the legend
plt.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
```
